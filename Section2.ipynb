```python
import numpy as np
from matplotlib import pyplot as plt
#import matplotlib as mpl
#mpl.rcParams.update(mpl.rcParamsDefault)

plt.figure(figsize=(7,5))
def task_1(x):
   return (abs(x-2)-1)**2

x = np.linspace(0, 5, 1000)

plt.plot(x, task_1(x), color='red')
plt.xlabel('Position of the target')
plt.ylabel('Cost Function value')
plt.title('Cost function ')
plt.show()
```

The cost function $(|x-2|-1)^2$, has two local minimums, one positioned at $x = 1$ and the other positioned at $x = 3$.

This is due to the fact that, for a range $r=1$, the target must be positioned at $a\pm r =  a \pm 1 $.
Therefore if $ a = 2 $, then in order to minimize the value of the cost function for a given $x$, the target must be positioned at either $x = 1$ or $x = 3$.

Between the values $x = 1$ and $x = 3$ there is a spike at $x = 2$, this is due to the increasing proximity of the target to the anchor (when we are between these two local minimums). Has such $|x-2| \rightarrow 0$ while $ r = 1$.

Finally, the further away the target is from these two local minimums, (i.e $x \rightarrow 0$ or $x \rightarrow +\inf$  ), the more the value of the cost function increases.
